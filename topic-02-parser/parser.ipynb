```python
def parse(s):  # return a number
    n = 0
    in_fraction = False
    is_negative = False
    d = 0.1
    exponent = 0
    exp_negative = False
    in_exponent = False
    
    if s[0] == "-":
        is_negative = True
        s = s[1:]
    
    for c in s:
        if c in "eE":
            in_exponent = True
        elif in_exponent:
            if c == "-":
                exp_negative = True
            elif c == "+":
                pass
            else:
                exponent = exponent * 10 + ord(c) - ord("0")
        elif c == ".":
            if not in_fraction:
                in_fraction = True
            else:
                raise Exception("syntax error, more than one '.'")
        elif not in_fraction:
            n = n * 10 + ord(c) - ord("0")
        else:
            n = n + (ord(c) - ord("0")) * d
            d = d / 10
    
    if exp_negative:
        exponent = -exponent
    
    result = n * (10 ** exponent)
    
    if is_negative:
        return -result
    else:
        return result
```

```python
# Helper function to check for floating-point equality
def close_to(x, y, delta=0.000001):
    return abs(x - y) < delta

# Test for single digit
def test_single_digit():
    assert parse("0") == 0
    assert parse("1") == 1
    assert parse("2") == 2
    assert parse("9") == 9

# Test for multiple digits
def test_multiple_digits():
    assert parse("10") == 10
    assert parse("123") == 123
    assert parse("00123") == 123
    
# Test for floating point numbers
def test_floating_point():
    assert close_to(parse("0.1"), 0.1)
    assert close_to(parse("123.456"), 123.456)
    assert close_to(parse(".789"), 0.789)
    assert close_to(parse("456."), 456.0)

def test_negative_integers():
    assert parse("-5") == -5
    assert parse("-48") == -48
    assert close_to(parse("-3.4"), -3.4)
    assert close_to(parse("-.8"), -0.8)
    assert close_to(parse("-3."), -3.0)

# Test for scientific notation
def test_scientific_notation():
    assert close_to(parse("1e2"), 100)
    assert close_to(parse("1e-2"), 0.01)
    assert close_to(parse("12.34e5"), 1234000)
    assert close_to(parse("12.34e-5"), 0.0001234)
    assert close_to(parse("12.34E5"), 1234000)
    assert close_to(parse("12.34E-5"), 0.0001234)


# Run all tests
test_single_digit()
test_multiple_digits()
test_floating_point()
test_negative_integers()
test_scientific_notation()

print("done")
```
